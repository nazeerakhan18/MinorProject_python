```python
import re
import numpy as np
from datetime import datetime
from collections import Counter
```

```python
with open("hostel_bois.txt", "r", encoding="utf-8") as file:
    chat_lines = file.readlines()

print("File Loaded Successfully")
print("Total Lines :", len(chat_lines))
```

```python
pattern = r'^(\d{2}/\d{2}/\d{2}), (\d{2}:\d{2}) - (.*?): (.*)$'

messages = []

for line in chat_lines:

    line = line.strip()

    match = re.match(pattern, line)

    if match:

        date = match.group(1)
        time = match.group(2)
        sender = match.group(3)
        message = match.group(4)

        messages.append({
            "date": date,
            "time": time,
            "sender": sender,
            "message": message
        })

print("Messages Parsed :", len(messages))
```

```python
participants = set()

message_count = Counter()

date_count = Counter()

hour_count = Counter()

for msg in messages:

    participants.add(msg["sender"])

    message_count[msg["sender"]] += 1

    date_count[msg["date"]] += 1

    hour = int(msg["time"].split(":")[0])

    hour_count[hour] += 1

print("Participants :", len(participants))

print("Total Messages :", len(messages))
```

```python
most_active = max(message_count, key=message_count.get)

most_active_day = max(date_count, key=date_count.get)

most_active_hour = max(hour_count, key=hour_count.get)

print("="*50)

print("GROUP OVERVIEW")

print("="*50)

print("Participants :", len(participants))

print("Messages :", len(messages))

print("Most Active Person :", most_active)

print("Most Active Day :", most_active_day)

print("Most Active Hour :", f"{most_active_hour}:00")
```

```python
all_text = ""

for msg in messages:

    all_text += " " + msg["message"].lower()

clean_text = re.sub(r"[^a-zA-Z\s]", "", all_text)

words = clean_text.split()

stop_words = {
"the","is","am","are","was","were","a","an","and","or","to","of",
"in","on","for","with","at","by","from","as","it","its","this",
"that","be","been","being","i","me","my","we","our","you","your",
"he","she","they","them","their","him","her","do","does","did",
"have","has","had","will","would","can","could","should","shall",
"if","but","so","because","not","yes","no","ok","okay","ya","yeah",
"im","ill","dont","cant","bro","bhai","guys","lol","haha"
}

filtered_words = []

for word in words:

    if word not in stop_words and len(word) > 2:

        filtered_words.append(word)

word_count = Counter(filtered_words)

print("="*50)

print("TOP 10 WORDS")

print("="*50)

for word,count in word_count.most_common(10):

    print(word,"-",count)
```

```python
activity_matrix = np.zeros((24,7),dtype=int)

for msg in messages:

    date_obj = datetime.strptime(msg["date"],"%d/%m/%y")

    day = date_obj.weekday()

    hour = int(msg["time"].split(":")[0])

    activity_matrix[hour][day] += 1

print("Activity Matrix")

print(activity_matrix)
```

```python
spammer = max(message_count,key=message_count.get)

ghost = min(message_count,key=message_count.get)

night_count = Counter()

for msg in messages:

    hour = int(msg["time"].split(":")[0])

    if hour >=23 or hour<5:

        night_count[msg["sender"]]+=1

night_owl = max(night_count,key=night_count.get) if night_count else "None"

print("="*60)

print("GROUP DNA REPORT")

print("="*60)

print("Total Messages :",len(messages))

print("Participants :",len(participants))

print("Most Active :",most_active)

print("Spammer :",spammer)

print("Ghost :",ghost)

print("Night Owl :",night_owl)

print("Most Active Day :",most_active_day)

print("Most Active Hour :",f"{most_active_hour}:00")

print("Top Word :",word_count.most_common(1)[0][0])

print("="*60)
```

```python
print("=" * 60)
print("MESSAGE SHARE")
print("=" * 60)

total = len(messages)

for person, count in message_count.most_common():

    percentage = (count / total) * 100

    print(f"{person:<15} {count:>4} messages   ({percentage:.2f}%)")
```

```python
print("=" * 60)
print("TOP 5 ACTIVE MEMBERS")
print("=" * 60)

rank = 1

for person, count in message_count.most_common(5):

    bar = "█" * (count // 5)

    print(f"{rank}. {person}")
    print(f"   Messages : {count}")
    print(f"   {bar}")
    print()

    rank += 1
```

```python
report = f"""
=============================
GROUP DNA REPORT
=============================

Total Messages : {len(messages)}

Participants : {len(participants)}

Most Active : {most_active}

Spammer : {spammer}

Ghost : {ghost}

Night Owl : {night_owl}

Most Active Day : {most_active_day}

Most Active Hour : {most_active_hour}:00

Top Word : {word_count.most_common(1)[0][0]}

=============================
"""

with open("Group_Report.txt", "w") as file:

    file.write(report)

print("Report Saved Successfully!")
```
